```python
def f(x):
    return 3*(x**2)-(4*x)
```

```python
def numerical_lim(f,x,h):
    return (f(x+h)-f(x))/h
```

```python
h = 0.1
for i in range(8):
    print(f'h = {h:.5f}, numerical limit = {numerical_lim(f,1,h):.5f}')
    h *=0.1
```
