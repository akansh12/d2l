### Saving Memory

```python
before = id(Y)
Y = Y+X
id(Y) == before
```

```python
Z = torch.zeros_like(Y)
print('ID:{}'.format(id(Z)))
Z[:] = X+Y
print('ID:{}'.format(id(Z)))
```

```python
before = id(X)
X +=Y
id(X) == before
```
